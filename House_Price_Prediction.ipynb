## Now, let's check the assumptions for Multiple Linear Regression

```python
# 1. Assumption: Linearity
# We'll plot actual vs predicted values to check for linearity
plt.scatter(y_test, y_pred)
plt.title("Actual vs Predicted Values")
plt.xlabel("Actual Median House Value")
plt.ylabel("Predicted Median House Value")
plt.show()
```

```python
# 2. Assumption: No Multicollinearity
# We'll use the Variance Inflation Factor (VIF) to check for multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm
# Add a constant term for VIF calculation
X_vif = sm.add_constant(X_train)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i+1) for i in range(len(X.columns))]  # +1 for the constant

print("\nVariance Inflation Factor (VIF):")
print(vif_data)
```

```python
#4. Assumption: Homoscedasticity (Equal variance of residuals)
# We'll plot residuals vs predicted values to check for homoscedasticity
residuals = y_test - y_pred
plt.scatter(y_pred, residuals)
plt.title("Residuals vs Predicted Values")
plt.xlabel("Predicted Values")
plt.ylabel("Residuals")
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```
